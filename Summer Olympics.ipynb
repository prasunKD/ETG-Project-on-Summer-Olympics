### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
#2
data = []
c=0
for md in df['Medal']:
    c+=1
    if md=='Gold':
        data.append(df.iloc[c]['Sport'])
s=set(data) 
ls=[]
for i in s:
    ls.append([data.count(i),i])
ls = sorted(ls)[-1:-6:-1]
x=[]
y=[]

for i in range(5):
  x.append(ls[i][1])
  y.append(ls[i][0])

plt.bar(x,y)
plt.plot()
```
